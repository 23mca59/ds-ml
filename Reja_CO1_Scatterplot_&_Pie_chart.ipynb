


#### Activity 1: Create scatter plot.



```
# 1. Create a scatter plot between the 'SepalLengthCm' & 'SepalWidthCm' columns & Differentiate between the data points of different classes using the 'hue' parameter.
import seaborn as sn
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/23mca59/ds-ml/main/Iris.csv')
df.head()
```

```
df.columns
```

```
sn.scatterplot(x=df['SepalLengthCm'],y=df['SepalWidthCm'],hue=df['Species'])
plt.title("Sepal length vs Sepal Width")
plt.show()
```

```
# 2. Create a scatter plot between the 'PetalLengthCm' & 'PetalWidthCm' columns & Differentiate between the data points of different classes using the 'hue' parameter.
sn.scatterplot(x=df['PetalLengthCm'],y=df['PetalWidthCm'],hue=df['Species'])
plt.title("Petal length vs Petal width")
plt.show()
```
